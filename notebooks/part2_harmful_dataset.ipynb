```python
import torch
import matplotlib.pyplot as plt
from torch import nn
from torchvision import datasets, transforms
from collections import OrderedDict
```

*CONECTAR EL DATASET*

Aqui definimos la ruta del data, hacemos transformaciones, cargamos el dataset y por medio DataLoader cargamos las imágenes en grupos (batches), en este caso de tamaño 32. Además, se utiliza la opción shuffle=True para mezclar los datos de forma aleatoria, evitando que el modelo aprenda patrones basados en el orden de las imágenes.

```python
#Ruta
data_dir = '../data/class'

#Transformaciones
transform = transforms.Compose([
    transforms.Resize((64, 64)), #ajusta el tamaño de la imagen 
    transforms.ToTensor()
])

#Cargar el Dataset
dataset = datasets.ImageFolder(data_dir, transform=transform)

dataloader = torch.utils.data.DataLoader(dataset, batch_size=32, shuffle=True)

#verificamos la clases 
print(dataset.classes)
```

*MOSTRAR UNA IMAGEN*

### Visualización de imágenes

Las imágenes se reorganizan utilizando la función permute(1,2,0), ya que PyTorch maneja el formato de las imágenes como (canales, alto, ancho), mientras que matplotlib requiere el formato (alto, ancho, canales).

```python
images, labels = next(iter(dataloader))

print(images.shape)
```

```python
plt.imshow(images[0].permute(1, 2, 0))
plt.title(f"Clase: {dataset.classes[labels[0]]}")
plt.show()
```

### Preparación de los datos

Antes de ingresar las imágenes a la red neuronal, se reorganizan en un vector unidimensional utilizando la función view. Esto es necesario porque las capas lineales (Linear) requieren datos en forma de vector.

### Forward Pass

Se realiza un forward pass, donde una imagen del dataset es procesada por la red neuronal. En esta etapa, el modelo genera probabilidades para cada una de las clases. Dado que la red no ha sido entrenada, estos valores son aleatorios.

```python
input_size = 64 * 64 * 3
hidden_sizes = [400, 200, 100]
output_size = 6

model = nn.Sequential(OrderedDict([
    ('fc1', nn.Linear(input_size, hidden_sizes[0])),
    ('relu1', nn.ReLU()),
    ('fc2', nn.Linear(hidden_sizes[0], hidden_sizes[1])),
    ('relu2', nn.ReLU()),
    ('fc3', nn.Linear(hidden_sizes[1], hidden_sizes[2])),
    ('relu3', nn.ReLU()),
    ('output', nn.Linear(hidden_sizes[2], output_size)),
    ('softmax', nn.Softmax(dim=1))
]))
```

### Resultado del modelo

Se obtuvo la salida de la red neuronal en forma de probabilidades para cada clase. Aunque el modelo aún no ha sido entrenado, se puede observar cómo procesa la información y genera una predicción inicial.

```python
images = images.view(images.shape[0], -1)
ps = model(images[0].unsqueeze(0))

print("Probabilidades:", ps)
print("Clase predicha:", dataset.classes[labels[0]])
```
