*CREAR TENSORES*

Un tensor es la base de todo: 
vector-1D
Matriz-2D
imageb-3D

En la guia los tensores son la base fundamental de las redes neuronales, creamos un tensor x de tamaño 3x2 con valores aleatorios entre 0 y 1.

```python
x = torch.rand(3,2)
print("Tensor x:") 
print(x)
```

*OPERACIONES*
las redes neuronales hacen sumas y multiplicaciones, se crea un tensor y del mismo tamaño que x, pero en este caso sera lleno de unos y tambien Se suman los tensores x | y de forma elemento a elemento.

```python
y= torch.ones (x.size())
print("Tensor y:")
print(y)
z= x + y 
print("Tensor z = x + y:")
print(z)
```

*INDEXACIÒN* 
Esto es importante porque segun la guia con esto podemos aceder a datos especificos, en donde el entrenamiento de la neurona: 

- sleccionar pixeles

```python
print(z[0])
print(z[:, 1])
```

*OPERACIONES IN-PLACE*

Usamos _ para modificar el tensor original y esimportante en optimizacion (memoria y velociidad)

```python
print("Sumando 1 a cada elemento de 'z':")
print("Actualizacion tensor z con z.add_(1)")
z.add_(1)
```

Las redes neuronales necesitan datos en formato específico y para eso usamos *RESHAPE*

Ejemplo:

Imagen: 20x20
Red: vector de 700

```python
print(" Modificamos 'z' a (2,3) con resize_:")
z.resize_(2, 3)
print(z)
```

*NUMPY ↔ TORCH*
Aqui realizamos la conversion entre Numpy y Torch, en la que se crea un array de NumPy 'a' de tamaño 4x3 con valores aleatorios, luego convierte el array de NumPy a un tensor de PyTorch 'b', despues se convierte 'b' nuevamente a un array de NumPy, multiplicamos y finalmente como 'b' y 'a' comparten la misma memoria, al modificar 'b' también se actualiza 'a'.

```python
a = np.random.rand(4,3)

b = torch.from_numpy(a)
print("Array de NumPy 'a' (4x3) con valores aleatorios")
print(a)

print("Convirtiendo 'a' a tensor de PyTorch 'b': ")
print(b)

b_new = b.numpy()
print("Se combierte 'b' nuevamente a array de NumPy (b_new):")
print(b_new)


b.mul_(2)
print("Tensor 'b' después de b.mul_(2):")
print(b)

print("Array 'a' actualizado:")
print(a)
```
